# 3. Misclassification analysis

```python
corpus = pd.read_csv("../corpora/german_modern_poems_epochpoet.csv")
```

## LSVM

```python
mpid = []
wrong_labels = {}

with open("../results/ml/misclassifications/pid_lsvm.json", "r") as f:
    lsvm_mis = json.load(f)
    for k, v in lsvm_mis.items(): 
        for k2, v2 in v.items():
            for v3 in v2:
                mpid.append(v3)
                wrong_labels[v3] = k2
```

```python
mcorpus = corpus[corpus.pid.isin(mpid)]  
del mcorpus["epoch_year"]
del mcorpus["filename"]
del mcorpus["title"]
del mcorpus["poemlength"]
mcorpus['wrong_epoch'] = mcorpus['pid'].map(wrong_labels)
```

```python
jahr_natur = mcorpus[(mcorpus.epoch_poet == "Jahrhundertwende") & (mcorpus.wrong_epoch == "Naturalismus")]

lsvm_natur_df = pd.DataFrame(jahr_natur.poet.value_counts()).rename_axis("Dichter").reset_index()
lsvm_natur_df.columns = ["Dichter", "Fehlklassifizierungen"]
poet_counts = dict(corpus.poet.value_counts())
lsvm_natur_df["Relative Hufigkeiten der Fehlklassifizierungen in %"] = np.around((lsvm_natur_df["Fehlklassifizierungen"] / lsvm_natur_df['Dichter'].map(poet_counts))*100, decimals=2)
lsvm_natur_df["Anteil an Gesamtzahl aller Gedichte im Korpus"] = np.around((lsvm_natur_df['Dichter'].map(poet_counts) / corpus.shape[0]) * 100, decimals=2)
```

### LSVM: Verteilung der falsch klassifizierten Gedichten nach Erscheinungsjahren, die flschlicherweise dem Naturalismus anstatt der Jahrhundertwende zugeordnet wurden 

```python
ax = sns.distplot(jahr_natur.year,
                        kde=False, 
                        rug=False, 
                        bins=45, 
                        hist_kws=dict(edgecolor="black", linewidth=1, alpha=0.75),
                        axlabel="Jahr")
ax.set_xticks(list(range(1880, 1910, 5)))
highest = sorted([p.get_height() for p in ax.patches])[-1]
shighest = sorted([p.get_height() for p in ax.patches])[-2:-1]
for p in ax.patches:
    if p.get_height() >= shighest and p.get_height() <= highest:
        p.set_color("purple")
        p.set_edgecolor("black")
        p.set_label("1887")
    if p.get_height() >= highest:
        p.set_color("crimson")
        p.set_edgecolor("black")
        p.set_label("1892")
ax.set_label("Erscheinungsjahre")
plt.title("LSVM: Jahrhundertwende  Naturalismus")
plt.legend()
ax.grid(False)
ax.set_facecolor("#F5F5F5")
plt.savefig("../resources/interim_results/lsvm_misclassification_publicationyear_jahr_natur.png", dpi=300)
plt.show()
```

## BERT

```python
bmpid = []
bwrong_labels = {}
bwrong_labels_alternative = {}

with open("../results/bert/misclassifications/pid_poetc_da_rede(27.06.20_15:43).json", "r") as f:
    bert_mis = json.load(f)
    bert_mis = bert_mis["epoch_poet"]
    
    for cv, cv_v in bert_mis.items():
        for k, v in cv_v.items():
            for k2, v2 in v.items():
                for pid in v2:
                    bmpid.append(pid)
                    if pid in bwrong_labels:
                        if bwrong_labels[pid] != k2:
                            bwrong_labels_alternative[pid] = k2
                    else:
                        bwrong_labels[pid] = k2
```

```python
bwrong_labels_alternative
```

```python
bmcorpus = corpus[corpus.pid.isin(bmpid)]  
del bmcorpus["epoch_year"]
del bmcorpus["filename"]
del bmcorpus["title"]
del bmcorpus["poemlength"]
bmcorpus['wrong_epoch'] = bmcorpus['pid'].map(bwrong_labels)
```

```python
bjahr_natur = bmcorpus[(bmcorpus.epoch_poet == "Jahrhundertwende") & (bmcorpus.wrong_epoch == "Naturalismus")]

bert_natur_df = pd.DataFrame(bjahr_natur.poet.value_counts()).rename_axis("Dichter").reset_index()
bert_natur_df.columns = ["Dichter", "False Negatives"]
poet_counts = dict(corpus.poet.value_counts())
bert_natur_df["Relative Hufigkeit der Fehlklassifizierungen in %"] = np.around((bert_natur_df["False Negatives"] / bert_natur_df['Dichter'].map(poet_counts))*100, decimals=2)
bert_natur_df["Anteil an Korpus in %"] = np.around((bert_natur_df['Dichter'].map(poet_counts) / corpus.shape[0]) * 100, decimals=2)
```

```python
def random_downsampling(corpus, class_col = "epoch_poet", max_value = 500):
    """ Reduces all instances of all classes to a certain maximum value.
    """   
    
    corpus_1 = corpus[corpus[class_col] == 'Expressionismus']
    corpus_2 = corpus[corpus[class_col] == 'Jahrhundertwende']
    corpus_3 = corpus[corpus[class_col] == 'Naturalismus']
    
    corpus_1 = corpus_1.sample(max_value)
    corpus_2 = corpus_2.sample(max_value)
    corpus_3 = corpus_3.sample(max_value)

    return pd.concat([corpus_1, corpus_2, corpus_3], axis=0)
```

```python
ep2 = random_downsampling(ep)
```
